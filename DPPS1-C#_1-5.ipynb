### Question 3 C#
##### Implement function triangleArea(a,b,c) that takes as input the lengths of the 3 sides of a triangle and returns the area of the triangle. By Heron's formula, the area of a triangle with side lengths a, b, and c is s(s - a)(s -b)(s -c) , where s = (a+b+c)/2. 


```
// function to calculate area of triangle using semiperimeter - sp
class areaTri
    {
        public double triangleArea(double a, double b, double c)
        {
            double sp = (a + b + c) / 2;  
            return Math.Sqrt(sp * ((sp - a) * (sp - b) * (sp - c)));  
        }
        
    }
           
```

```
// calling the function for area of triangle with given inputs
double a = 2.0;
double b=2.0;
double c=2.0;
areaTri areat = new areaTri();
double ar ;
ar =  areat.triangleArea(a,b,c);
Console.WriteLine("Value after calling the function {0}",ar);
```

### Question 4 C#

##### Write a program in C# Sharp to separate odd and even integers in separate arrays.
##### Go to the editor
##### Test Data :
##### Input the number of elements to be stored in the array :5
##### Input 5 elements in the array :
##### element - 0 : 25
##### element - 1 : 47
##### element - 2 : 42
##### element - 3 : 56
##### element - 4 : 32
##### Expected Output:
##### The Even elements are:
##### 42 56 32
##### The Odd elements are :
##### 25 47

```
// As Jupyter Notebook doesn't support the readLine function, I've use visual studio for solving further problems of C#

using System;
    
namespace DPpset1
{
    class Program
    {
        static void Main(string[] args)
        {
                int num;
                Console.Write("Input the number of elements to be stored in the array :");
                // Reading the number of elements and accordingly declare array of that size                 
                num = Convert.ToInt32(Console.ReadLine());
                
                //Declare array of the size of given input 
                int[] input_array = new int[num];
                int[] even_array = new int[num];
                int[] odd_array = new int[num];
                int i, j = 0, k = 0;


                //Read all the elements of the array                 
                Console.Write("Input {0} elements in the array :\n", num);
                for (i = 0; i < num; i++)
                {
                    Console.Write("element - {0} : ", i);
                    input_array[i] = Convert.ToInt32(Console.ReadLine());
                }
            
                //Traverse all elements of array to differentiate even and odd numbers in 2 different arrays             
                for (i = 0; i < num; i++)
                {
                    if (input_array[i] % 2 == 0)
                    {
                    //if the current element is even then store it in even array                         
                        even_array[j] = input_array[i];
                        j++;
                    }
                    else
                    {
                    //if the current element is odd then store it in odd array                                                  
                        odd_array[k] = input_array[i];
                        k++;
                    }
                }

                //Print even and odd arrays                 
                Console.WriteLine("\nThe Even array is : \n");
                for (i = 0; i < j; i++)
                {
                    Console.WriteLine("{0} ", even_array[i]);
                }

                Console.WriteLine("\nThe Odd odd array is :\n");
                for (i = 0; i < k; i++)
                {
                    Console.WriteLine("{0} ", odd_array[i]);
                }
                Console.WriteLine("\n\n");
            
            Console.ReadLine();

        }
    }
}
```

##### a. Write a function inside(x,y,x1,y1,x2,y2) that returns True or False depending on whether the point (x,y) lies in the rectangle with lower left corner (x1,y1) and upper right corner (x2,y2).
#### >>> inside(1,1,0,0,2,3)
#### True
#### >>> inside(-1,-1,0,0,2,3)
#### False


```
using System;

namespace DPpset1

{
    class Program
    {
        //Boolean function which checks whether the given points resides inside the rectangle or not    
        static bool inside(double x, double y, double x1, double y1, double x2,
                     double y2)
        {

            if ((x >= x1 && x <= x2) && (y >= y1 && y <= y2))
                return true;
            else
                return false;

        }
        static void Main(string[] args)
        {

            //Input the co-ordinates of points 
            double x, y, x1, y1, x2, y2;

            Console.WriteLine("Enter x : ");
            x = Convert.ToDouble(Console.ReadLine());
            Console.WriteLine("Enter y : ");
            y = Convert.ToDouble(Console.ReadLine());
            Console.WriteLine("Enter x1 : ");
            x1 = Convert.ToDouble(Console.ReadLine());
            Console.WriteLine("Enter y1 : ");
            y1 = Convert.ToDouble(Console.ReadLine());
            Console.WriteLine("Enter x2 : ");
            x2 = Convert.ToDouble(Console.ReadLine());
            Console.WriteLine("Enter y2 : ");
            y2 = Convert.ToDouble(Console.ReadLine());
            //call the function to calculate whether point is in rectangle
            Console.WriteLine(inside(x, y, x1, y1, x2, y2));
            Console.ReadLine();

        }
    }
}
```
